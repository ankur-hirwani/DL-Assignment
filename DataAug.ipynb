```python
##Importing the libraries requried to perform the Data Augmentation
from keras.preprocessing import image
import keras.utils as imag
from keras.preprocessing.image import ImageDataGenerator
```

```python
##Loading the data to be augmented.
img = imag.load_img('leaf.jpg',target_size=(300,400))
```

```python
##This is the data that will be augmented using various augmentation techniques
img
```

```python
##The ImageDataGenertor function of the Keras library take varios inputs on the type of augmentation to be performed and stores in an array.
##This array can be further converted to PIL.image type, and can be fed to the model. This eliminates the scarcity of the Training/Testing Data.
##The below code uses various techniques such as rotation,zoom-in/out,brightness etc. To produce various copies of data.
datagen = ImageDataGenerator(
rotation_range=0.5,
shear_range=0.2,
zoom_range=0.2,
horizontal_flip=True,
vertical_flip=True,
width_shift_range=0.2,
height_shift_range=0.2,
fill_mode='nearest',
brightness_range=(0,1)
 )
```

```python
##This line converts the array to ndarray format.
img = imag.img_to_array(img)
```

```python
##The ndarray is reshaped to a single image of 300*400 dimension
input_batch = img.reshape(1,300,400,3)
```

```python
##The given data is duplicated accorfing to the output given by the Image Data Generator function and stored in the AugData directory
i = 0

for output in datagen.flow(input_batch,batch_size=1,save_to_dir = 'AugData'):
    i = i+1
    if i==25:
        break;
    
```
